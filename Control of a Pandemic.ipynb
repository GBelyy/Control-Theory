Рассматривается задача борьбы с пандемией. Количество (часть от всего населения) инфицированных людей задается функцией $x(t) \in [0,1]$, где $t \in [0,T]$. При этом задается ограниченное управление (общественная политика лечения), определяемая как $u(t) \in [0,u^*]$.
При этом динамика системы описывается следующими дифференциальными уравнениями:
$$\dot{x} =  \alpha\cdot{(1-x)}\cdot{x} - u\cdot{x}, \
x (0) = x_0,$$
\
где $\alpha$ > 0 - некоторый фиксированный параметр, $x_0 \in [0,1]$ - некоторое фиксированное начальное число зараженных.

Требуется найти управление, при котором максимизируется следующий функционал (функция общественного благосостояния):
$$ J = -\int\limits_{0}^{T}e^{-rt}\cdot({x(t) + c\cdot{u(t)^k})}dt = \int\limits_{0}^{T} F dt → {\underset{0 \leq u \leq u^*}{max}},$$
\
где фиксированные коэффициенты r > 0, c > 0, k >=1 описывают изменения благосостояния.

Перейдем от задачи Лагранжа к задаче Майера, когда функционал представляет собой функцию $Ф(x(t_f), t_f)$.
\
Для этого введем новую переменную y:
$$ \dot{y} = F(x,u,t) = -e^{-rt}\cdot({x(t) + c\cdot{u(t)^k})}, y(0) = 0 $$

Тогда получим новую систему уравнений:
$$\begin{equation*}
 \begin{cases}
   \dot{x} =  \alpha\cdot{(1-x)}\cdot{x} - u\cdot{x}, 
   \\
  \dot{y} = -e^{-rt}\cdot({x(t) + c\cdot{u(t)^k})}.
 \end{cases}
\end{equation*}$$

При этом начальное многообразие имеет вид:
\begin{equation*} 
R(t_0=0) \in A = 
 \begin{cases}
  x(0) = x_0,
   \\
  y(0) = 0.
   \end{cases}
\end{equation*}   

Функционал примет вид:
$$ J = Ф(x(t_f), t_f) = y(T) → {\underset{0 \leq u \leq u^*}{max}} $$

Таким образом рассматривается задача оптимального управления со свободным правым концом и зафиксированным временем.

Функция Гамильтона-Понтрягина:
$$ H = p_x \cdot x + p_y \cdot y = p_x\cdot{(\alpha(1-x)\cdot{x} - u \cdot{x})} - p_ y \cdot e^{-rt}\cdot{(x+c\cdot{u^k})}$$
\
Тогда уравнения на сопряженнуе переменные примут вид:

$$\begin{equation*}
 \begin{cases}
     \dot{p_x} = - \frac{\partial H}{\partial x} = e^{-rt} - p_x\cdot{(\alpha - 2\alpha\cdot{x} - u)}
     \\
     \dot{p_y} = - \frac{\partial H}{\partial y} = 0
     \end{cases}
\end{equation*}$$

При этому условия трансверсальности на левом конце описываются следующими уравнениями:
$$ P(T) = \frac{\partial Ф}{\partial \begin{pmatrix} x\\ y\\t \end{pmatrix}} + \lambda_{t_f}  \frac{\partial t}{\partial \begin{pmatrix} x\\ y\\t \end{pmatrix}} = \begin{pmatrix} 0\\ 1\\\lambda_{t_f} \end{pmatrix},$$
\
где вектор фазовых переменных:
$$ P = \begin{pmatrix} p_x\\ p_y\\-H \end{pmatrix}.$$

Так как $p_y = const$ и из условия трансверсальности $p_y(T) = 1$, то $p_y = 1$ (далее везде будем осуществлять замену $p_y = 1$).

При этом на управление действует ограничение:
$$ u \leq u^*$$

Применим принцип максимума Понтрягина для двух случаев $k \in \{1,2\}$:
* $k=1$: 
$$H = p_x\cdot{(\alpha(1-x)\cdot{x} - u \cdot{x})} - e^{-rt}\cdot{(x+c\cdot{u})} = u ( -x\cdot{p} - ce^{-rt}) + p_x\alpha(1-x)\cdot{x} - e^{-rt}\cdot{x}$$
\
$$ \begin{equation*}
u = 
 \begin{cases}
   u^* &, p_x\cdot{x} < -ce^{-rt}\\
   0 & ,p_x\cdot{x} > -ce^{-rt}
 \end{cases}
\end{equation*}$$

* $k=2$: 
$$H = p_x\cdot{(\alpha(1-x)\cdot{x} - u \cdot{x})} - e^{-rt}\cdot{(x+c\cdot{u^2})}$$
\
При этом условие максимума гамильтониана дает соотношение:
$$ -p_x \cdot u \cdot{x} - e^{-rt}\cdot c\cdot{u^2} → max $$
\
Тогда оптимальное управление находится из условия эксремальности гамильтониана:
$$ \frac{\partial H}{\partial u} = -p_x \cdot x - 2 c e ^{-rt} \cdot u = 0 $$
\
Получаем, что:
$$\begin{equation*}
 \begin{cases}
   u = u^* , -\frac{p_x \cdot x} {2 \cdot c} e ^{rt} \geq u_* 
   \\
   u = -\frac{p_x \cdot x} {2 \cdot c} e ^{rt}, 0 < -\frac{p_x \cdot x} {2 \cdot c} e ^{rt} < u_*
   \\
   u = 0 , -\frac{p_x \cdot x} {2 \cdot c} e ^{rt} \leq 0
 \end{cases}
\end{equation*}$$
\
Переключение с одного режима на другой происходит на линиях:
$X_+$: $-\frac{p_x \cdot x} {2 \cdot c} e ^{rt} = u_*$  и $X_-$: $-\frac{p_x \cdot x} {2 \cdot c} e ^{rt} = 0 $

Источники, откуда были получены коэффициенты:
1) r - размер социальной скидки - https://www.researchgate.net/publication/255593139_A_Social_Discount_Rate_for_the_US \
2) alpha - некоторый коэффициент заболеваемости - https://en.wikipedia.org/wiki/COVID-19_pandemic_in_the_United_States \
3) с - процент затрат на здравоохранение - https://en.wikipedia.org/wiki/COVID-19_pandemic_in_the_United_States

```python
import numpy as np

from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

c = 0.16 #c >0 #
alpha = 0.001 # u_max > alpha
r = 0.4 #r > 0
u_max = 1
u_min = 0
T = 40
```

```python
def control_1(x,p,t):
    if p * x <= -c*np.exp(-r*t):
        return u_max
    else:
        return u_min

def control_2(x,p,t):
    if -(p * x)* np.exp(r*t)/(2*c) >= u_max:
        return u_max
    if -(p * x)* np.exp(r*t)/(2*c) < u_max and -(p * x)* np.exp(r*t)/(2*c) > 0:
        return -(p * x)* np.exp(r*t)/(2*c)
    if -(p * x)* np.exp(r*t)/(2*c) <= u_min:
        return u_min
    
def rhs_1(t, R):
    dRdt = np.empty(2)
    x = R[0].copy()
    p = R[1].copy()

    dRdt[0] = alpha * (1-x)*x - control_1(x,p,t)*x
    dRdt[1] = np.exp(-r*t) - p * (alpha - 2*alpha*x - control_1(x,p,t))

    return dRdt

def rhs_2(t, R):
    dRdt = np.empty(2)
    
    x = R[0].copy()
    p = R[1].copy()

    dRdt[0] = alpha * (1-x)*x - control_2(x,p,t)*x
    dRdt[1] = np.exp(-r*t) - p * (alpha - 2*alpha*x - control_2(x,p,t))

    return dRdt

def Ham_1(x,p,t):
    return p*(alpha*(1-x)*x - control_1(x,p,t)*x) - np.exp(-r*t)*(x + c * control_1(x,p,t))
def Ham_2(x,p,t):
    return p*(alpha*(1-x)*x - control_2(x,p,t)*x) - np.exp(-r*t)*(x + c * control_2(x,p,t)**2)
```

```python
def plotter(sol, option):
    k = int(option)
    
    x = sol.y[0].copy()
    p = sol.y[1].copy()
    t = sol.t.copy()
    
    Hams = np.empty(len(x),dtype = float)
    u = np.empty(len(x),dtype = float)
    if k == 1:
        for i in range(len(x)):
            Hams[i] = Ham_1(x[i],p[i],t[i])
            u[i] = control_1(x[i],p[i],t[i])
    else:
        for i in range(len(x)):
            Hams[i] = Ham_2(x[i],p[i],t[i])
            u[i] = control_2(x[i],p[i],t[i])
        
    plt.plot(t,x*100,'-r')
    plt.ylim([0,x0*1.1*100])
    plt.xlabel("время,усл. ед.")
    plt.ylabel("часть зараженных, %")
    plt.title(f'Зависимость количества инфицированных от времени в случае {k}')
    plt.grid()
    plt.show()
         
    plt.plot(t,u,'-b')
    plt.xlabel("время,усл. ед.")
    plt.ylabel("u, безразм. ед.")
    plt.title(f'Управление лечением от времени в случае {k}')
    plt.grid()
    plt.show()
    
    plt.plot(t,Hams,color = 'black')
    plt.xlabel("время,усл. ед.")
    plt.ylabel("H, безразм. ед.")
    plt.title(f'Гамильтониан от времени в случае {k}')
    plt.grid()
    plt.show()
```

```python
#начальные условия 
x0 = 0.01 # изначально заражен 1% населения


#начальное предположения для нахождения p(0)
init_1 = -100
init_2 = -100
```

```python
def objective_1(unk):
    sol = solve_ivp(rhs_1, [0,T], [x0, unk])
    p = sol.y[1]
    x = sol.y[0]
    return p[-1]
#используем условие трансверсальности на p_x

def objective_2(unk):
    sol = solve_ivp(rhs_2, [0,T], [x0, unk])
    p = sol.y[1]
    x = sol.y[0]
    return p[-1]
#используем условие трансверсальности на p_x

res_1 = fsolve(objective_1, init_1)
print('В первом случае p(0) = ',*res_1)
res_2 = fsolve(objective_2, init_2)
print('Во втором случае p(0) = ',*res_2)
```

```python
R0_1 = [x0,res_1]
R0_2 = [x0,res_2]

sol_1 = solve_ivp(rhs_1, [0, T], R0_1, method='LSODA')
sol_2 = solve_ivp(rhs_2, [0, T], R0_2, method='LSODA')
print(sol_1.message)
print(sol_2.message)
```

```python
plotter(sol_1,1)
```

```python
plotter(sol_2,2)
```
