Рассматривается задача задача перемещения КА из одной точки на заданной круговой орбите в другую точку на той же орбите, отличающуюся только фазовым углом $\phi$. При этом рассматриваются решения с минимальным временем.

Описывается движение космического корабля с точечной постоянной массой (тяга малая), движущегося в плоскости вокруг сферического гравитирующего центра (Земли) и управляемого постоянной малой тягой с изменяемым управлением.


Движение спутника в орбитальной плоскости описывается дифференциальными уравнениями (звездами обозначены размерные параметры):
\begin{equation*}
 \begin{cases}
  \dot{x^*} = v^*_x,
   \\
   \dot{y^*} = v^*_y,
   \\
   \dot{v^*_x} = -\frac{\mu^*\cdot{x^*}}{r^{*3}} + \frac{T^*}{m^*} \cos{\psi},
   \\
   \dot{v^*_y} = -\frac{\mu^*\cdot{y^*}}{r^{*3}} + \frac{T^*}{m^*} \sin{\psi} ,
   \end{cases}
\end{equation*}
\
где $x^*$,$y^*$ - компоненты радиус-вектора в орбитальной плоскости, $r^*$ = $\sqrt{x^{*2}+y^{*2}}$, $\mu^*$ - гравитационный параметр притягивающего тела (Земли), $\psi$ - угол направления тяги, отсчитываемый от направления оси X в инерцаильных координатах , $T^*$ - тяга двигателя, $m^*$ - масса КА.


Чтобы уравнения не менялись в зависимости от начальной орбиты и других Н.У. имеет смысл перейти к безразмерным параметрам. Воспользуемся формулами:
$$ \mu = \frac{DU^3}{TU^2},$$
где 1 DU = большая полуось орбиты, 1 TU = $\frac{период}{2\pi}$.
\
$$ T = \frac{T^* a^{*2}}{m^*\mu^*}$$
Тогда в безразмерном виде уравнения примут вид:
\begin{equation*}
 \begin{cases}
  \dot{x} = v_x,
   \\
   \dot{y} = v_y,
   \\
   \dot{v_x} = -\frac{\mu^\cdot{x}}{r^{3}} + T \cdot{\cos{\psi}},
   \\
   \dot{v_y} = -\frac{\mu^\cdot{y}}{r^{3}} + T \cdot{\sin{\psi}} ,
  \end{cases}
\end{equation*}


Пусть начальное многообразие определяется условиями:
\begin{equation*} 
R(t_0=0) \in A = 
 \begin{cases}
  x = 1,
   \\
  y = 0,
   \\
  v_x = 0,
    \\
  v_y = 1,
   \end{cases}
\end{equation*}
\
где $\vec{R}$ - вектор состояния в новых полярных координатах $\vec{R}$ = $[x,y,v_x,v_y]^T$.

Условия на правом конце должны удовлетворять нескольким условиям. Во - первых, КА должен оставаться на той же круговой орбите или $ x^2 + y ^2$ = $1$, $v_x^2 + v_y^2$ = $1$ и  $x\cdot{v_x} + y\cdot{v_y}$ = $0$ (условие ортогональности вектора скорости и радиус- вектора). Эти условия определяют терминальное многообразие:
\begin{equation*} 
X(t_f) \in B = 
 \begin{cases}
  x = \cos{(\phi + t_f)},
   \\
  y = \sin{(\phi + t_f)},
   \\
  v_x=-\sin{(\phi + t_f)},
   \\
  v_y = \cos{(\phi + t_f)},
   \end{cases}
\end{equation*}

ввиду безразмерности рассматриваемой системы $t_f$ - это угол, под которым КА перемещается во время маневра.

Требуется найти закон изменения управления (функцию $\psi (t)$),чтобы  обеспечить минимальное время достижения заданного фазового угла $t_f$:
$$Ф(x(t_f),t_f) = t_f → {\underset{\psi}{min}} $$
или в интегральном виде (приведем к стандартному виду задачи оптимизации, где функционал максимизируется):
$$ G(x(t_f),t_f) = -\int\limits_{0}^{t_f}1d\tau → {\underset{\psi}{max}} $$

Функция Гамильтона-Понтрягина:
$$ H = p_x\cdot{v_x} + p_y\cdot{v_y} + p_{v_x}(-\frac{x}{r^3} + T\cos{\psi}) + p_{v_y}(-\frac{y}{r^3} + T\sin{\psi})$$
Условие экстремума:
$$ \frac{\partial H}{\partial \psi} = -p_{v_x}\cdot{T\sin{\psi}} + p_{v_y}\cdot{T\cos{\psi}} = 0 $$

Тогда функция угла направления тяги $\psi$ определяется из следующего равенства:
$$\psi = \tan^{-1}{\frac{p_{v_y}}{p_{v_x}}} $$

Перейдем к сопряженным переменным:
\begin{equation*}
 \begin{cases}
  \dot{p_x} = \frac{p_{v_x}}{r^3} - \frac{3\cdot{x}}{r^5}\cdot{(x\cdot{p_{v_x}} + y\cdot{p_{v_y}})},
   \\
  \dot{p_y} = \frac{p_{v_y}}{r^3} - \frac{3\cdot{y}}{r^5}\cdot{(x\cdot{p_{v_x}} + y\cdot{p_{v_y}})},
   \\
   \dot{p_{v_x}} = -p_x,
   \\
   \dot{p_{v_y}} = -p_y,
   \end{cases}
\end{equation*}


При этом необходимо определить начальные условия для сопряженных переменных
($\href{https://www.researchgate.net/publication/2621013_Application_of_Energy_Storage_To_Solar_Electric_Propulsion_Orbital_Transfer/citations}{статья}$):
\
Поскольку гамильтониан линеен по своим по сопряженным переменным, то зададим произвольно условие: $ p_x(0) = 1 $;
\
При этом остаются неизвестными еще три начальных условия для сопряженных переменных. Для их нахождения воспользуемся терминальными условиями из множества $B$. Тогда задачу их нахождения можно поставить следующим образом:
\
\
$$ F(p_y(0),p_{v_x}(0),p_{v_y}(0),t_f) = \begin{bmatrix}
  x(t_f) - \cos{(\phi + t_f)} \\
  y(t_f) - \sin{(\phi + t_f)} \\
  p_{v_x}(t_f) + \sin{(\phi + t_f)} \\
  p_{v_y}(t_f) - \cos{(\phi + t_f)} 
\end{bmatrix} = 0$$
\
Эту задачу можно решить методом пристрелки. Начальные приближения для нее берутся из $\href{https://www.researchgate.net/publication/266385507_Minimum-Time_Orbital_Phasing_Maneuvers}{статьи}$.

В данном примере рассмотрены следующие начальные условия:
1) $x^*(0)$ = $R_e$ + 500 км = 6871 км, где $R_e$ - радиус Земли = 6371 км - рассматривается маневр на низкой околоземной орбите
\
2) $v_y^*(0)$ = $\sqrt{G\frac{M}{x(0)}}$ = 7.615 $\frac{км}{сек}$
\
3) $T^*$ = 42 $Н$
\
4) $m^*$ = 100 кг
\
5) $\phi$ = 0.2101 рад ~ 12 градусов

Также рассматривается угол $\theta$ - угол между тягой и локальной вертикалью КА. Тогда $\theta$ определяется через полученные угол между тягой и инерциальной осью OX $\phi$ и полярный угол места КА $\gamma$, определяемый как угол между инерциальной осью OX и радиус-вектором спутника, с помощью выражения:
$$ \theta = \psi - \gamma - \frac{\pi}{2}$$

```python
import numpy as np
from numpy import cos, sin, tan
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from scipy.optimize import fsolve

T = 0.05
phi = 0.2101
mu = 1 
```

```python
def control(pv):
  return np.arctan(pv[1]/pv[0])

def rhs(t, R, options):
    if options == 'controled': #интегратор для траектории с управлением
        T_u = T
    if options == 'uncontroled': #интегратор для траектории без управления
        T_u = 0

    dRdt = np.empty((8,))

    rvec = R[0:2].copy()
    vvec = R[2:4].copy()
    p_rvec = R[4:6].copy()
    p_vvec = R[6:8].copy()

    r = np.linalg.norm(rvec)
    phase_vec = np.array([np.cos(control(p_vvec)),np.sin(control(p_vvec))])

    dRdt[0:2] = vvec
    dRdt[2:4] = -(mu/r**3) * rvec + T_u * phase_vec
    dRdt[4:6] = p_vvec/r**3 - (3* rvec/r**5)* np.dot(rvec, p_vvec)
    dRdt[6:8] = -p_rvec
    
    return dRdt

def Hamiltonian(R):
    x = R[0]
    y = R[1]
    v_x = R[2]
    v_y = R[3]
    p_x = R[4]
    p_y = R[5]
    p_v_x = R[6]
    p_v_y = R[7]
    
    psi = np.arctan(p_v_y/p_v_x)
    r = np.sqrt(x**2 + y **2)
    return 1 + p_x*v_x + p_y*v_y + p_v_x*(-x/r**3 + T*np.cos(psi)) + p_v_y*(-y/r**3 + T*np.sin(psi))
```

```python
def plotter(traj, orbit):
    x = orbit.y[0]
    y = orbit.y[1]

    x_traj = traj.y[0]
    y_traj = traj.y[1]
    v_x_traj = traj.y[2]
    v_y_traj = traj.y[3]
    p_x_traj = traj.y[4]
    p_y_traj = traj.y[5]
    p_v_x_traj = traj.y[6]
    p_v_y_traj = traj.y[7]

    psi = np.arctan2(p_v_y_traj,p_v_x_traj)
    psi_ship = (np.arctan(p_v_y_traj/p_v_x_traj) - np.arctan(y_traj/x_traj) - np.pi/2)
    
    H = Hamiltonian([x_traj,y_traj,v_x_traj,v_y_traj,p_x_traj,p_y_traj,p_v_x_traj,p_v_y_traj])
    
    
    plt.figure(figsize = (8,8))
    axes = plt.gca()
    
    for i in range(len(x_traj)):
        x0 = x_traj[i]
        y0 = y_traj[i]

        x1 = 0.05*np.cos(psi[i]) + x0
        y1 = 0.05*np.sin(psi[i]) + y0
        if i == 0:
            line = Line2D([x0, x1], [y0, y1], color="black",label = 'Направление тяги')
        else:
            line = Line2D([x0, x1], [y0, y1], color="black")
        axes.add_line(line)
        
    plt.plot(x_traj,y_traj,color = 'red',label = 'Траектория с управлением',linewidth = 2)
    plt.plot(x,y,'--b',label = 'Исходная орбита')
    plt.title(f'Траектория КА при T={T}, $%s$ = {phi}'%'\phi')
    plt.legend()
    plt.grid()
    axes.set_aspect("equal")
    print('\n *Черные линии обозначают направление управляющей тяги')
    plt.show()
    
    plt.plot(sol_traj.t, np.rad2deg(psi_ship))
    plt.xlabel('t, TU')
    plt.ylabel('theta, градусы')
    plt.title('Угол управляющей тяги относительно локальной вертикали корабля')
    plt.grid(True)
    plt.show()

    plt.plot(sol_traj.t, H)
    plt.xlabel('t, TU')
    plt.ylabel('H, ND')
    plt.title('Гамильтониан')
    plt.grid(True)
   
    plt.show()
```

```python
#для построения
option  = ['controled','uncontroled']

#начальные приближения
init_cond = [0.123,0.6906,0.5000,3.6599]
```

```python
#Метод пристрелки

def objective(unk):
    sol = solve_ivp(lambda t, R: rhs(t, R, option[0]), [0, unk[3]],[1, 0, 0, 1, 1, unk[0], unk[1], unk[2]])
    x = sol.y[0]
    y = sol.y[1]
    v_x = sol.y[2]
    v_y = sol.y[3]
    return [x[-1] - np.cos(phi + unk[3]),
            y[-1] - np.sin(phi + unk[3]),
            v_x[-1] + np.sin(phi + unk[3]),
            v_y[-1] - np.cos(phi + unk[3])] #используем терминальные условия 

res = fsolve(objective, init_cond)
print('Найденные начальные условия [p_y(0),p_vx(0),p_vy(0),tf] \n',*res)
```

```python
# Вектор времени
t0 = 0
tf = res[3] #в радианах

tol = 1e-12

# начальные условия
R0 = np.array([1, 0, 0, 1, 1, res[0], res[1], res[2]])
option  = ['controled','uncontroled']

sol_traj = solve_ivp(lambda t, R: rhs(t, R, option[0]), [t0, tf], R0, method='LSODA', atol = tol, rtol = tol)
sol_orbit = solve_ivp(lambda t, R: rhs(t, R, option[1]), [t0, 2*np.pi], R0, method='LSODA', atol = tol, rtol = tol)
print(sol_traj.message)

plotter(sol_traj, sol_orbit)
```
